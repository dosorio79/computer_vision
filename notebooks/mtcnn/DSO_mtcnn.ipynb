```python
from mtcnn import MTCNN
import cv2
import os
from glob import glob
```

```python
folder_list = glob(os.path.join("../../raw_data/Kaggle_clean/", '*'))
#print(filename_list)

filename_list = []
for folder in folder_list:
    filename_list.append(glob(os.path.join(folder, "*.png")))
```

```python
filename_list
```

```python
def crop_faces(img, image, faces):
    if len(faces) > 0:
        for i, face in enumerate(faces):
            x, y, w, h = face['box']
            
            # get face and save crop
            face = image[y:y + h, x:x + w]
            face = cv2.cvtColor(face, cv2.COLOR_RGB2BGR)
            cv2.imwrite(f"{img.strip('.png')}_face{i}.png", face)
            print(f"{img.strip('.png')}_face{i}.png is saved")
    return None   
```

```python
img.strip('.png').replace('Kaggle_clean', 'cropped')
```

```python
detector = MTCNN()

for img_path in filename_list:
    for i, img in enumerate(img_path):
        image = cv2.cvtColor(cv2.imread(img), cv2.COLOR_BGR2RGB)
        #Outputs a faces list of dict, with the bounding box inside the key 'box'
        faces = detector.detect_faces(image)
        crop_faces(img, image, faces)
```

# Single file approach

```python
img_path = 'img/'
img_file = 'lateral.jpg'
img_type = img_file.split('.')[1]
image = cv2.cvtColor(cv2.imread(f'{img_path}{img_file}'), cv2.COLOR_BGR2RGB)
#Outputs a faces list of dict, with the bounding box inside the key 'box'
detector = MTCNN()
faces = detector.detect_faces(image)
```

```python
if len(faces) > 0:
    for i, face in enumerate(faces):
        x, y, w, h = face['box']
        
        # get face and save crop
        face = image[y:y + h, x:x + w] 
        cv2.imwrite(f'{img_path}face{i}.{img_type}', face)
        print(f"face{i}.{img_type} is saved")    
        cv2.imshow("Cropped Face", face)
        
        # draw rectangle
        color = (0, 255, 255) # in BGR
        stroke = 1
        cv2.rectangle(image, (x, y), (x + w, y + h), color, stroke)

# Show both images, press any key to exit     
cv2.imshow('Face detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
    
```
